```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("BTC_FINAL.csv")
df['Date']= pd.to_datetime(df['Date'])
df.info()
```

```python
#Make a new varibel for the log of the BTC_Price
df['BTC_Price_log'] = np.log(df['BTC_Price'])

#Make a new varibel with the difference in the day to day price:
df['Daily_change'] = df.BTC_Price_log.diff(periods=1)
#solve the problem with the missing values in the row 1:

#no columns contain any missing value = very happy! 
```

```python
list1 = df["BTC_Price"]
list2 = df["BTC_Price_log"]

plt.figure(figsize = (28,14))

plt.subplot(1, 2, 1)
plt.hist(list1,bins = 7)
plt.title('BTC Price',fontsize=40)
plt.tick_params(axis='x', labelsize=30)
plt.tick_params(axis='y', labelsize=35)
plt.xlabel('USD',fontsize=30)

plt.subplot(1, 2, 2)
plt.hist(list2,bins = 7)
plt.title('BTC Price (log)',fontsize=40)
plt.tick_params(axis='x', labelsize=34)
plt.tick_params(axis='y', labelsize=35)
plt.xlabel('USD (log)',fontsize=30)

plt.show()
```

```python
plt.rcParams['figure.figsize'] = (12.0, 6.0)
prices = pd.DataFrame({"Price (USD)":df["BTC_Price"], "Log Price (USD)":df["BTC_Price_log"]})
prices.hist()
plt.tick_params(axis='x', labelsize=22)
plt.tick_params(axis='y', labelsize=22)
plt.show()
```

```python
high_corr = ['BTC_Price','GOOGLET_SELL','REDDIT_SCORE','REDDIT_COMMENT',
                      'WIKI_HIST_EN','WIKI_HITS_ZH','WIKI_HITS_JA','WIKI_HITS_FR','WIKI_HITS_GE','WIKI_HITS_RU',
                      'WIKI_HITS_KO','WIKI_HITS_PE','WIKI_HITS_HA','WIKI_HITS_ID','WIKI_HITS_VI','GOOGLET_CRYPTO',
            'GOOGLET_BUY','TWEET_BUY', 'TWEET_SELL','WIKI_WARNING', 'WIKI_OTHER', 'WIKI_NEGATIVE', 'WIKI_REGULATION']
plt.rcParams['font.size'] = 14

# Heast map most correlated


pearsoncorr = df[high_corr].corr(method='pearson')

hmap = sns.heatmap(pearsoncorr, xticklabels=True, yticklabels=True, vmax=1, center=0, linewidths=.1,
            square=True, cbar_kws={"shrink": .8}, annot_kws={"size":10}, annot=True, fmt='.2f', cmap='vlag')
sns.despine()
hmap.figure.set_size_inches(22,22)
plt.show()
```
