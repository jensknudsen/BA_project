```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew
```

```python
df = pd.read_csv("BTC_FINAL.csv")
df['Date']= pd.to_datetime(df['Date'])
df
```

```python
pearsoncorr = df.corr(method='pearson')

hmap = sns.heatmap(pearsoncorr, xticklabels=True, yticklabels=True, vmax=1, center=0, linewidths=.1,
            square=True, cbar_kws={"shrink": .8}, annot_kws={"size":6.5}, annot=True, fmt='.2f', cmap='vlag')
sns.despine()
hmap.figure.set_size_inches(22,22)
plt.show()
```
