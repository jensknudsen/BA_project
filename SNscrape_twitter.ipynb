```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
import time
```

```python
start = time.time()
# Creating list to append tweet data to
tweetslist = []

# Using TwitterSearchScraper to scrape data cointaining the words 'Bitcoin' and 'Buy' and append tweets to list
for i,tweet in enumerate(sntwitter.TwitterSearchScraper('Bitcoin Buy since:2015-07-01 until:2021-09-10').get_items()):
    if i>400000000:
        break
    tweetslist.append([tweet.date, tweet.id, tweet.content, tweet.user.username, tweet.likeCount])
    
# Creating a dataframe from the tweets list above
tweets = pd.DataFrame(tweetslist, columns=['Datetime', 'Tweet Id', 'Text', 'Username','Like'])
end = time.time()
print(end - start)
```

```python
tweets.to_csv('Tweets_Buy_1.csv', sep=',', index=False)
```

```python
tweets['Datetime'] = pd.to_datetime(tweets['Datetime']).dt.date
tweets = tweets.groupby('Datetime')[['Like']].sum()
idx = pd.date_range('3-29-2010', '9-10-2021')
tweets.index = pd.DatetimeIndex(tweets.index)
tweets = tweets.reindex(idx, fill_value=0)
tweets = pd.DataFrame(tweets)
tweets = tweets.reset_index()
tweets = tweets.rename(columns={'index': 'Date', 'Like': 'Value'})
tweets
```

```python
# Export dataframe into a CSV
tweets.to_csv('Tweets_Buy.csv', sep=',', index=False)
```
