```python
! pip install attrdict
from attrdict import AttrDict
import pageviewapi
import pandas as pd
```

```python
#ENGLISH WIKI
```

```python
d = pageviewapi.aggregate('en.wikipedia', '2015070100', '2021091000', access='all-access',
                      agent='all-agents', granularity='daily')

for key in d:
    print(d[key])
df = d[key]
df = pd.DataFrame(df)
df = df.drop('project', inplace=False, axis=1)
df = df.drop('access', inplace=False, axis=1)
df = df.drop('agent', inplace=False, axis=1)
df = df.drop('granularity', inplace=False, axis=1)
df['timestamp'] = df['timestamp'].astype(str).str[:-2]
df['timestamp'] = pd.to_datetime(
                          df['timestamp'],
                          format='%Y%m%d'
)
df.to_csv('WIKI_HITS_EN.csv', sep=',', index=False)
```
