```python
import pandas as pd # library for data analysis
import requests # library to handle requests
from bs4 import BeautifulSoup # library to parse HTML documents
```

```python
# parse data from the html into a beautifulsoup object
response = requests.get(r'https://timelines.issarice.com/wiki/Timeline_of_Bitcoin#cite_note-428')
soup = BeautifulSoup(response.text, 'html.parser')
table = (soup
         .find('h2', text='Full timeline')
         .find_next_sibling('table'))
```

```python
df=pd.read_html(str(table))
# convert list to dataframe
df=pd.DataFrame(df[0])
# remove the rows where month and date is equal to NaN:
df = df[df['Month and date'].notna()]
df = df[df['Month and date'].str.contains('1|2|3|4|5|6|7|8|9|0', na=False)]
# Combine the year, month and date into onw column 
cols = ["Month and date", "Year"]
df['Date'] = df[cols].apply(lambda row: ' '.join(row.values.astype(str)), axis=1)
df["Date"].replace({"December 9–10 2002": "December 9 2002", "October 9–12 2009": "October 9 2009"}, inplace=True)
#Make the datetime right:
import datetime as datetime
import datetime
listen = df["Date"].tolist()

dates = []

for element in listen:
    date = datetime.datetime.strptime(element, '%B %d %Y').strftime('%d/%m/%Y')
    dates.append(date)
    
df['Year'] = dates

df = df.drop(['Month and date','Date'], axis = 1)
df = df.rename(columns={'Year': 'Date'})
df = df.reset_index(drop = True)
df
```

```python
df_negative = df[df['Event type'].str.contains('(negative)', regex=False, case=False, na=False)]
df_regulation = df[df['Event type'].str.contains('(regulation)', regex=False, case=False, na=False)]
df_warning = df[df['Event type'].str.contains('(warning)', regex=False, case=False, na=False)]
df_other = df[~df["Event type"].str.contains('warning', na=False)]
df_other = df_other[~df_other["Event type"].str.contains('negative', na=False)]
df_other = df_other[~df_other["Event type"].str.contains('regulation', na=False)]

df_negative = df_negative.drop_duplicates(subset=['Date'], keep='first')
df_regulation = df_regulation.drop_duplicates(subset=['Date'], keep='first')
df_warning = df_warning.drop_duplicates(subset=['Date'], keep='first')
df_other = df_other.drop_duplicates(subset=['Date'], keep='first')
```

```python
df_other.to_csv('df_other.csv', sep=',', index=False)
df_regulation.to_csv('df_regulation.csv', sep=',', index=False)
df_warning.to_csv('df_warning.csv', sep=',', index=False)
df_negative.to_csv('df_negative.csv', sep=',', index=False)
```
