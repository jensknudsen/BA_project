```python
from pytrends.request import TrendReq
from pytrends import dailydata

df = dailydata.get_daily_data('bitcoin buy', 2010, 3, 2021, 9)

print(df)
df = df.reset_index()
df.to_csv('googletrend_buy.csv', sep=',', index=False)
```

```python
df.tail(30)
```
